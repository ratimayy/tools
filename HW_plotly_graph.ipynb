Data from: https://www.goodcarbadcar.net/2024-us-auto-sales-figures-by-brand-brand-rankings/

```python
import pandas as pd
import plotly.express as px
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/ratimayy/tools/main/car_fixed.csv')
df.head()
```

```python
df.info()
```

```python
a = df[(df['Brand']=='Ford') | (df['Brand']=='Toyota') | (df['Brand']=='Chevrolet') | (df['Brand']=='Honda') | (df['Brand']=='Nissan')]
a
```

```python
a.drop('Q1 2024', axis = 1, inplace = True)
a.drop('Q2 2021', axis = 1, inplace = True)
a.drop('Q3 2021', axis = 1, inplace = True)
a.drop('Q4 2021', axis = 1, inplace = True)

a
```

```python
import plotly.express as px

# Melt the DataFrame to convert it to long format
df_long = a.melt(id_vars=["Brand"], var_name="Quarter", value_name="Sales")

# Plot the bar chart using Plotly Express
fig = px.bar(df_long, x="Brand", y="Sales", color="Quarter",
             title="Top 5 Car Sales by Brands from Quarters1 2022 to Quarter4 2023 in US",
             labels={"Sales": "Number of Sales", "Brand": "Car Brand", "Quarter": "Quarter"})

# Set the Y-axis range
fig.update_yaxes(range=[0, 4000000])

# Show the plot
fig.show()
```
